```python
# Importando a biblioteca pandas no Python

import pandas as pd

# Importando banco de dados e criando dataframe
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
#Criando um novo dataframe com as colunas escolhidas
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]

display(ca_df)
```

### Função dataframe.loc[]

```python
#Exibe a 4ª linha
display(ca_df.loc[3])
```

```python
#Exibe da 10ª a 20ª linha
display(ca_df.loc[9:19])
```

```python
#Cria um dataframe gas_df coontendo apenas 4 colunas
#Revenda, Municipio, Produto, Valor de Venda)
#Somente com combustível sendo GASOLINA
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']

display(gas_df)
display(gas_df['Valor de Venda'].max())
```

```python
# Exibir qual o posto e município da gasolina mais cara do país

display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

# Aula 3

Data: 08/02/2023

```python
# Dataframe.loc[] com múltiplas condições para filtrar os dados
# Qual são os postos e os preços de ETANOL na minha cidade?
# Ordenado do menor para o maior valor de venda

etanol_canoas_df = ca_df.loc[(ca_df['Produto']=='ETANOL') & (ca_df['Municipio']=='CANOAS')]
display(etanol_canoas_df.sort_values(by='Valor de Venda'))
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA
#do Bairro Mooca em SÃO PAULO?

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | 
                             (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA
#do Bairro Mooca em SÃO PAULO?
#função isin()

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(['GASOLINA', 'GASOLINA ADITIVADA'])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda por tipo de combustível:
media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)

display(media_por_combustivel_df)
```

```python
# Quero adicionar uma coluna de valor booleano chamada "Ativo"
#que, por padrão, vai ser True para todas as linhas

combustiveis_df['Ativo'] = True

print(combustiveis_df.info())
display(combustiveis_df.head(3))
```

```python
# E se eu quiser uma coluna nova chamada "Obs"
#que preencha automaticamente "MELHOR CIDADE"
#quando o Município for CANOAS, RS

combustiveis_df['Obs'] = ["Melhor Cidade" if municipio == 'Canoas' else '' for municipio in combustiveis_df['Municipio']]
```

```python
# Exportar um dataframe para Excel

etanol_canoas_df.to_excel('etanol_canoas.xlsx', sheet_name='Etanol em Canoas', index=False)
```
