## Segundo exemplo de manipulação de dados usando Python Pandas

**Inserção de dados, remoção de colunas, merge, gráficos...**

```python
# Importando biblioteca pandas no Python
import pandas as pd
```

```python
# Importando banco de dados e criando dataframe
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Inserção simples de dados
combustiveis_df['Ativo'] = True

display(combustiveis_df)
```

```python
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE"
#quando a coluna Municipio for igual a SAO PAULO

combustiveis_df['Obs'] = ["Melhor Cidade" if municipio == 'SAO PAULO' else None for municipio in combustiveis_df['Municipio']]
display (combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO', 'CAMPINAS']), ['Municipio', 'Obs']])
```

```python
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE"
#quando a coluna Municipio for igual a SAO PAULO
# versão Numpy

import numpy as np

combustiveis_df['Status do Valor de Venda']=np.where(combustiveis_df['Valor de Venda']> 6.5, 'Caro', 'Barato')

display(combustiveis_df[['Revenda','Valor de Venda', 'Status do Valor de Venda']])
```

```python
# Calcular postos de gasolina por habitantes temos na
# amostragem de combustíveis nov/2021

num_hab_df = pd.read_csv('ibge_num_habitantes_estimado.csv', sep=';')
num_hab_df.rename(columns={'Estado':'Estado - Sigla'}, inplace=True)
display(num_hab_df)
```

```python
# Merge dos dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_hab_df, how='inner', on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
# Função dropna() para destruir colunas vazias

merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
#Função drop() para se livrar de colunas
colunas=['Regiao - Sigla', 'Nome da Rua', 'Bairro', 'Cep', 'Produto',
         'Data da Coleta', 'Valor de Venda', 'Unidade de Medida', 'Bandeira',
         'Numero Rua', 'Ativo', 'Status do Valor de Venda']
merge_df.drop(axis='columns', inplace=True, labels=colunas)

print(merge_df.info())
```

```python
#Função drop_duplicate()
merge_df.drop_duplicates(inplace=True)
display(merge_df)
```

```python
# Agrupar e contar quantos postos tem na cidade... 

postos_por_mun_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_mun_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_mun_df.rename(columns={'Revenda': 'Num_postos'}, inplace=True)
display(postos_por_mun_df)
```

```python
#Posto por habitante
postos_por_mun_df['NumHabitantesPorPosto']=postos_por_mun_df['NumHabitantes2021']/postos_por_mun_df['Num_postos']

display(postos_por_mun_df)
```

```python
# Criando gráficos
# Importando biblioteca matplotlib

import matplotlib.pyplot as plt

plt.hist(combustiveis_df['Valor de Venda'])
# Vamos colocar um título e rótulos nos eixos do gráfico
plt.title('Preço dos combustíveis - Nov/2021')
plt.xlabel('Preço em R$')
plt.ylabel('Quantidade de Coletas')

#Traça a linha vermelha tracejada com o preço médio
plt.axvline(combustiveis_df['Valor de Venda'].mean(), color='red',
            linestyle='dashed', linewidth=2)

# Exibe o gráfico independente da IDE
plt.show()
```

```python
#Segundo gráfico
#Visualização do consumo médio
c_mean = combustiveis_df['Valor de Venda'].groupby(by=combustiveis_df['Produto']).mean()
display(c_mean)
```

```python
import seaborn as sns
# Definir a área do gráfico
plt.figure(figsize=(10,6))

#Plotar o gráfico
c_mean_grafico = c_mean.plot(
    kind="barh",
    title= "Média Preço por Combustiveis",
    xlabel="Tipo de Combustivel",
    ylabel="Preço reais/litro",
    color="red",
    alpha = 0.9,
)

c_mean_grafico.set_ylabel("Tipo de Combustivel")
c_mean_grafico.set_xlabel("Preço reais/litro")



#Grid
plt.grid(axis='x')
#Remove as linhas superior e lateral direita do gráfico
sns.despine()
#Exibe
plt.show()
```

```python
display(c_mean)

excel = 'por_litro.xlsx'

c_mean.to_excel(excel, 'Sumário')
```

```python
#Vamor personalizar a planilha Excel...
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment

#Abrir o excel no openpyxl
wb = load_workbook(excel)

ws = wb['Sumário']

#Vamos pintar o cabeçalho da tabela de "cinzinha"
cinzinha=PatternFill('solid',fgColor='A0A0A0')
coords = ['A1', 'B1']
for coord in coords:
  ws[coord].fill = cinzinha

#Onde o preço do combustível for maior ou igual a 6,5 reais pinta a
#fonte de vermelho e deixa negrito
MAX_ROW = ws.max_row
num_linha=2
while(num_linha <= ws.max_row):
  coord = 'B'+str(num_linha)
  if ws[coord].value >= 6.5:
    ws[coord].font = Font(bold=True, color='FF0000')
  else:
    ws[coord].font = Font(color='00FF00')
  num_linha += 1


#Salvar o excel
wb.save(excel)
```
