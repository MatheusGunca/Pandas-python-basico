## Segundo exemplo de manipulação de dados usando Python Pandas

**Inserção de dados, remoção de colunas, merge, gráficos...**

```python
# Importando biblioteca pandas no Python
import pandas as pd
```

```python
# Importando banco de dados e criando dataframe
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Inserção simples de dados
combustiveis_df['Ativo'] = True

display(combustiveis_df)
```

```python
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE"
#quando a coluna Municipio for igual a SAO PAULO

combustiveis_df['Obs'] = ["Melhor Cidade" if municipio == 'SAO PAULO' else None for municipio in combustiveis_df['Municipio']]
display (combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO', 'CAMPINAS']), ['Municipio', 'Obs']])
```

```python
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE"
#quando a coluna Municipio for igual a SAO PAULO
# versão Numpy

import numpy as np

combustiveis_df['Status do Valor de Venda']=np.where(combustiveis_df['Valor de Venda']> 6.5, 'Caro', 'Barato')

display(combustiveis_df[['Revenda','Valor de Venda', 'Status do Valor de Venda']])
```

```python
# Calcular postos de gasolina por habitantes temos na
# amostragem de combustíveis nov/2021

num_hab_df = pd.read_csv('ibge_num_habitantes_estimado.csv', sep=';')
num_hab_df.rename(columns={'Estado':'Estado - Sigla'}, inplace=True)
display(num_hab_df)
```

```python
# Merge dos dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_hab_df, how='inner', on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
# Função dropna() para destruir colunas vazias

merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
#Função drop() para se livrar de colunas
colunas=['Regiao - Sigla', 'Nome da Rua', 'Bairro', 'Cep', 'Produto',
         'Data da Coleta', 'Valor de Venda', 'Unidade de Medida', 'Bandeira',
         'Numero Rua', 'Ativo', 'Valor de Venda - Status', 
         'Status do Valor de Venda']
merge_df.drop(axis='columns', inplace=True, labels=colunas)

print(merge_df.info())
```

```python
#Função drop_duplicate()
merge_df.drop_duplicates(inplace=True)
display(merge_df)
```

```python
# Agrupar e contar quantos postos tem na cidade... 

postos_por_mun_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_mun_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_mun_df.rename(columns={'Revenda': 'Num_postos'}, inplace=True)
display(postos_por_mun_df)
```

```python
#Posto por habitante
postos_por_mun_df.reset_index(inplace=True)

postos_por_mun_df['Postos por Habitante'] = postos_por_mun_df['Num_postos']/postos_por_mun_df['NumHabitantes2021']

display(postos_por_mun_df)
```
